```python
df = pd.read_csv("../dataset/raw_dataset.csv")
df
```

```python
df.style.set_properties(subset=['content'], **{'width': '300px'})
```

# Case Folding

Menyeragamkan huruf pada kalimat

```python
case_folding = []
for review in df['content']:
    # Ubah menjadi Lowercase
    case_folding.append(review.lower())
case_folding[1]
```

# Proses Data Cleaning (Menghapus karakter non-ASCII)

```python
proses_cleaning1 = []
for clean in case_folding:
    proses = clean.encode("ASCII", "ignore")
    proses_cleaning1.append(proses.decode())
proses_cleaning1[38]
```

# Proses Data Cleaning (menghapus angka)

```python
from string import digits
proses_cleaning2 = []
for clean in proses_cleaning1:
    proses = str.maketrans('', '', digits)
    proses_cleaning2.append(clean.translate(proses))
proses_cleaning2[3]
```

# Proses Data Cleaning (menghapus tanda baca)

```python
proses_cleaning3 = []
for clean in proses_cleaning2:
    proses_cleaning3.append(clean.translate(str.maketrans('', '', string.punctuation)))
proses_cleaning3[30]
```

# Proses Data Cleaning (menghapus spasi yang berlebihan)

```python
proses_cleaning4 = []
for clean in proses_cleaning3:
    proses_cleaning4.append(re.sub(r'\s+', ' ', clean, flags=re.I))
proses_cleaning4[40]
```

# Proses Data Cleaning (menghapus 1 karakter)

```python
proses_cleaning5 = []
for clean in proses_cleaning4:
    proses_cleaning5.append(re.sub(r'\s+[a-zA-Z]\s+', ' ', clean))
proses_cleaning5[40]
```

# Proses Data Cleaning (menghapus 1 karakter dari awal)

```python
proses_cleaning6 = []

for clean in proses_cleaning5:
    proses_cleaning6.append(re.sub(r'\^[a-zA-Z]\s+', ' ', clean))
proses_cleaning6
```

```python
data_clean = pd.DataFrame(proses_cleaning6, columns=['text'])
data_clean[50:60]
```

```python
data_clean.to_csv("../dataset/proses_cleaning6.csv", index=None)
```

```python
hasil_normalisasi = pd.read_csv("../dataset/hasil_normalisasi.csv")
hasil_normalisasi
```

```python
hasil_normalisasi.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
hasil_normalisasi.columns = ['text']
```

```python
hasil_normalisasi = hasil_normalisasi.iloc[1:5000]
```

```python
hasil_normalisasi
```

# Stemming (Proses Lama)

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemming_proses = []

for stem_ind in hasil_normalisasi['text']:
    stemming_proses.append(stemmer.stem(stem_ind))
stemming_proses[30]
```

# FILTERING / STOPWORD REMOVAL (PROSES LAMA)

```python
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
stopword_proses = []

for kalimat in stemming_proses:
    stopword_proses.append(stopword.remove(kalimat))
stopword_proses
```

# Tokenize

```python
tokenize_proses = []

for kalimat in stopword_proses:
    tokenize_proses.append(nltk.tokenize.word_tokenize(kalimat))
tokenize_proses
```

# Menggabungkan Hasil Tokenize 

```python
gabung_tokenize = []

for kata in tokenize_proses:
    gabung_tokenize.append(' '.join(kata))
gabung_tokenize
```

```python
dataset_bersih = pd.DataFrame(gabung_tokenize, columns=['text'])
dataset_bersih
```

# Menghapus data kosong

```python
dataset_bersih = dataset_bersih[dataset_bersih.text != '']
```
